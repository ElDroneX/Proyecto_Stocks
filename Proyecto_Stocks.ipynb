```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import yfinance as yf
from plotly.subplots import make_subplots
import plotly.graph_objects as go
import warnings
warnings.filterwarnings("ignore", category=FutureWarning)
```

```python
print("\nPregunta 1\n")
tesla_data = yf.download("TSLA", period="5y")
tesla_data.reset_index(inplace=True)
print(tesla_data.head(), "\n")
```

```python
print("\nPregunta 2\n")
url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0220EN-SkillsNetwork/labs/project/revenue.htm"
html = requests.get(url).text
soup = BeautifulSoup(html, "html.parser")

tesla_revenue = pd.DataFrame(columns=["Date", "Revenue"])
for row in soup.find_all("table")[0].find("tbody").find_all("tr"):
    col = row.find_all("td")
    date = col[0].text
    revenue = col[1].text
    tesla_revenue = pd.concat([
        tesla_revenue,
        pd.DataFrame({"Date":[date], "Revenue":[revenue]})
    ], ignore_index=True)
print(tesla_revenue.tail(), "\n")
```

```python
print("\nPregunta 3\n")
gme_data = yf.download("GME", period="5y")
gme_data.reset_index(inplace=True)
print(gme_data.head(), "\n")
```

```python
print("\nPregunta 4\n")
gme_revenue = pd.DataFrame(columns=["Date", "Revenue"])
for row in soup.find_all("table")[1].find("tbody").find_all("tr"):
    col = row.find_all("td")
    date = col[0].text
    revenue = col[1].text
    gme_revenue = pd.concat([
        gme_revenue,
        pd.DataFrame({"Date":[date], "Revenue":[revenue]})
    ], ignore_index=True)
print(gme_revenue.tail(), "\n")
```

```python
print("\nPregunta 5\n")
tesla_revenue["Revenue"] = tesla_revenue["Revenue"].str.replace(r"[\$,]", "", regex=True)
tesla_revenue = tesla_revenue[tesla_revenue["Revenue"] != ""]
tesla_revenue["Revenue"] = pd.to_numeric(tesla_revenue["Revenue"], errors="coerce")
tesla_revenue = tesla_revenue.dropna()

fig1 = make_subplots(rows=2, cols=1, shared_xaxes=True,
                     subplot_titles=("Tesla Share Price", "Tesla Revenue"))
fig1.add_trace(go.Scatter(x=tesla_data["Date"], y=tesla_data["Close"], name="Share Price"), row=1, col=1)
fig1.add_trace(go.Bar(x=tesla_revenue["Date"], y=tesla_revenue["Revenue"], name="Revenue"), row=2, col=1)
fig1.update_layout(height=800)
fig1.show()
```

```python
print("\nPregunta 6\n")
gme_revenue["Revenue"] = gme_revenue["Revenue"].str.replace(r"[\$,]", "", regex=True)
gme_revenue = gme_revenue[gme_revenue["Revenue"] != ""]
gme_revenue["Revenue"] = pd.to_numeric(gme_revenue["Revenue"], errors="coerce")
gme_revenue = gme_revenue.dropna()

fig2 = make_subplots(rows=2, cols=1, shared_xaxes=True,
                     subplot_titles=("GameStop Share Price", "GameStop Revenue"))
fig2.add_trace(go.Scatter(x=gme_data["Date"], y=gme_data["Close"], name="Share Price"), row=1, col=1)
fig2.add_trace(go.Bar(x=gme_revenue["Date"], y=gme_revenue["Revenue"], name="Revenue"), row=2, col=1)
fig2.update_layout(height=800)
fig2.show()
```
